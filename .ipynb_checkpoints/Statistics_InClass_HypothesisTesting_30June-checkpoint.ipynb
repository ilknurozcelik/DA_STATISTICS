## Significance Tests
Örneklem parametrelerini kullanarak popülasyon parametlerrini tahmin etmek için kullanılan Anlamlılık Testi(Significance test) dir.

## Hypothesis

popülasyon parametresi hakkında bir iddia da bulunmak. bir cümle yazmak. bu hipotezdir.
popülasyon parametreleri ortalama, oran, vb olabilir.

hipotezin amacı  populasyonun parametreleri hakkında ortaya atılan bir iddianın dogru olup olmadıgını sample verilerini kullanarak kontrol etmektir.


## Steps of a Significance Test

1. Assumptions
2. Hypotheses
3. Test Statistics
4. P-Value
5. Conclusion


### Step 1: Assumptions
Önce değişkenlerinizi ve parametreleri tanımlarız.
Uygun varsayımlar varsa o testi yaparsınız. Yani varsayımlarınız hangi testi yapmanız için uygun ise o testi (z veya t testi) yaparsınız.

Z-test assumptions (z score kullanılarak yapılır):
1. Örnekler random seçilmeli
2. Gözlemler birbirinden bağımsız olmalı (ör: 30 farklı kişi)
3. Popülasyon std. sapması biliniyor veya en az 30 örnekleminiz olmalı

### Step2 : Hypothesis

The null hypothesis : parametrenin belirli bir değer aldığı durumu ifade eder.
Null Hipotezi : H0 olarak gösterilir : başlangıçta doğru olduğu kabul edilen değerdir.

The alternative hypothesis : parametrenin alternatif bir değer aralığına düştüğünü ifade eder.
Alternatif Hipotez: H1 olarak gösterilir. H0'a aykırı bir iddiadır.

Bunlar birbirini dışlar ve genellikle ayrıntılıdır.

Bir hipotez testinin muhtemel sonuçları:

1. H0 hipotezini reddetmek (reject H0)
2. H0 hipotezini reddetmeyi başaramamak (fail to reject H0)

## Step 3: Test Statistics
Aşağıdaki şemaya göre hangi testi yapacağımıza karar verebiliriz. Ve belirlediğimiz teste göre test istatistiklerini hesaplarız.




## Step 4: P-Value

Test istatistiğinden hesaplanır. Bir olasılık değeridir.

P-value ne kadar küçük olursa sample istatistiğiniz 0 hipotezini reddetmek için o kadar güçlü bir kanıt sunmuş olur. Böylece  null hipotezi reddedebilirsiniz.

p-value için eşik değer alpha:significance level dır. p-value alpha ile karşılaştırılır.


## Step 5 : Conclusion

Hesaplanan p değerine göre (yani p değerini alpha değeri ile karşılaştırarak) null hipotezi reject edip edemeyeceğimize karar veririz.

p-value > alpha : fail to reject the null hypothesis
p-value < alpha : reject the null hypothesis


## Significance Level

Önem/Anlamlılık düzeyi (alfa veya α), sıfır hipotezini reddetmeden ve etkinin istatistiksel olarak anlamlı olduğu sonucuna varmadan önce örnekte bulunacak kanıtın gücünün bir ölçüsüdür.

P değeri anlamlılık düzeyi(alpha veya α) ile karşılaştırılarak H0 hipotezinin reject edilip edilemeyeceğine karar verilir.

## Type I Error
Ör: Suçsuz olan birine suçlu dersek tip 1 hata yapmış oluruz.
Null hipotez doğru iken reddederseniz Tip 1 hata olur.

## Type II Error

Ör: Suçlu olan birisine suçsuz dersek tip 2 hata yapmış oluruz.
Null hipotez yanlış iken kabul ederseniz Tip 2 hata olur.






## One Tailed - Two Tailed Test

İki kuyruklu:  Bazen Reddetme Bölgesi iki parçaya bölünür ve test istatistiğinin örnekleme dağılımının her iki kuyruğunda bulunur (H 0 :μ = μ 0 doğru olduğunda)

H 0 : "İlacın etkisi yoktur (μ = μ 0 )" 

H 1 : "İlacın etkisi var (μ ≠ μ 0 )"


Tek kuyruklu:  Bazen Reddetme Bölgesi, test istatistiğinin örnekleme dağılımının bir kuyruğunda yer alır (H 0 :μ = μ 0 , H 1 :μ > μ 0'a karşı doğru olduğunda ).

H 0 : "İlacın etkisi yoktur (μ = μ 0 )" 

H 1 : "İlaç tepki süresini azaltır (μ < μ 0 )"

# One Sample t Test

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## Test Statistics
Tek kuyruk hipotez testi

```python
xbar=10.5  # population mean
sigma= 1.5  # population standart deviation
n= 40  # sample size
mu = 10  # sample mean
```

#### Hypothesis

H0 : mu = 10

H1 : mu > 10

```python
# test statistic

z = (xbar -mu) /(sigma/np.sqrt(n))   # (sigma/np.sqrt(n)) is standart error
z
```

## p-value

```python
p = 1 - stats.norm.cdf(z)
p
```

```python
# Elinizde array şeklinde veri varsa doğrudan aşağıdaki metodu kullanarak da p value hesaplanabilir.

#  stats.ttest_1samp(a, popmean, alternative="greater")
```

```python
df= sns.load_dataset("mpg")
```

```python
df.head()
```

```python
# P-Value from a Z-Score in Python

# stats.norm.sf(abs(z))
```

```python
df[df["origin"] == "usa"].describe()
```

H0 : mu = 110  usa arabalarının ortalama horsepower değerleri 110'dur.
H1 : mu > 110  usa arabalarının ortalama horsepower değerleri 110' dan büyüktür.

```python
stats.ttest_1samp(df[df["origin"] == "usa"]["horsepower"], 110, alternative="greater")
```

```python
stats.ttest_1samp(df[df["origin"] == "usa"]["horsepower"].dropna(), 110, alternative="greater")
```

H0 : mu = 22  usa arabalarının ortalama mpg değerleri 22'dir.
H1 : mu < 22  usa arabalarının ortalama mpg değerleri 22' den küçüktür.

```python
stats.ttest_1samp(df[df["origin"] == "usa"]["mpg"], 22, alternative="less")
```

```python
df.origin.value_counts()
```

```python
df[df["origin"] == "europe"].describe()
```

H0 : mu = 100  europe arabalarının ortalama horsepower değerleri 100'dur.
H1 : mu > 100  europe arabalarının ortalama horsepower değerleri 110' dan büyüktür.

```python
stats.ttest_1samp(df[df["origin"] == "europe"]["horsepower"].dropna(), 100, alternative="greater")

# H1' de mu 100'den büyüktür dediğimiz için alternative parametresine "greater" yazdık.
```

H0 : mu = 3000  europe arabalarının ortalama weight değerleri 3000'dir.
H1 : mu != 3000  europe arabalarının ortalama weight değerleri 3000 değildir (two-sided).

```python
stats.ttest_1samp(df[df["origin"] == "europe"]["weight"].dropna(), 3000, alternative="two-sided")
```

H0 : mu = 3000  europe arabalarının ortalama weight değerleri 3000'dir.
H1 : mu < 3000  europe arabalarının ortalama weight değerleri 3000 'den küçüktür.

```python
stats.ttest_1samp(df[df["origin"] == "europe"]["weight"].dropna(), 3000, alternative="less")

# H1' de mu 3000'den küçüktür dediğimiz için alternative parametresine "less" yazdık.
```
